```python
# Merge on City column
merged_df = pd.merge(df1, df2, on="City", how="outer")

# Save merged data
merged_df.to_csv("merged_crime_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt

# Plot trend of rape cases over years
plt.figure(figsize=(10,5))
df1.groupby("City")[['2020', '2021', '2022']].sum().plot(kind="line", marker="o")
plt.title("Trend of Rape Cases Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Cases")
plt.grid()
plt.show()
```

```python
import seaborn as sns

top_cities = df2.sort_values("Total Rape", ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(data=top_cities, x="Total Rape", y="City", palette="Reds_r")
plt.title("Top 10 Cities with Highest Rape Cases")
plt.xlabel("Total Rape Cases")
plt.ylabel("City")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df1, x="Population", y="Crime_Rate", hue="City", size="Crime_Rate", sizes=(20, 200))
plt.title("Crime Rate vs Population")
plt.xlabel("Population (in Lakhs)")
plt.ylabel("Crime Rate per Lakh")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=['2020', '2021', '2022'], 
             y=df1[['2020', '2021', '2022']].sum(), marker="o")
plt.title("Rape Cases Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Rape Cases")
plt.grid()
plt.show()
```

```python
df_sorted = df2.sort_values(by="Total Rape", ascending=False).head(10)

plt.figure(figsize=(12, 6))
df_sorted[["Total Rape", 'Kidnapping cases']].plot(kind='bar', stacked=True, figsize=(12,6))
plt.title("Comparison of Rape vs Kidnapping Cases (Top 10 Cities)")
plt.xlabel("City")
plt.ylabel("Number of Cases")
plt.xticks(ticks=range(10), labels=df_sorted["City"], rotation=45)
plt.legend(["Total Rape", "Kidnapping cases"])
plt.show()
```

```python
age_groups = ["18 yr or above women rape", "Below 18 yr women rape"]
values = df2[age_groups].sum()

plt.figure(figsize=(8, 8))
plt.pie(values, labels=age_groups, autopct="%1.1f%%", colors=["blue", "orange"])
plt.title("Age Distribution of Rape Victims")
plt.show()
```
